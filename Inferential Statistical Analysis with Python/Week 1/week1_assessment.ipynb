#### Challenge problem:  
Write a function that take in four arguments: lst1, lst2, str1, str2, and returns a pandas DataFrame that has the first column labeled str1 and the second column labaled str2, that have values lst1 and lst2 scaled to be between 0 and 1.

For example
```
lst1 = [1, 2, 3]
lst2 = [2, 4, 5]
str1 = 'one'
str2 = 'two'

my_function(lst1, lst2, str1, str2)
``` 
should return a DataFrame that looks like:



|  <i></i> | one | two |
| --- | --- | --- |
| 0 | 0 | 0 |
| 1 | .5 | .666 |
| 2 | 1 | 1 |



````python
import numpy as np
import pandas as pd

def get_df(lst1, lst2, str1, str2):
    l1 = np.array(lst1) - min(lst1)
    l2 = np.array(lst2) - min(lst2)
    return pd.DataFrame({
        str1: l1 / max(l1),
        str2: l2 / max(l2),
    })

lst1 = [1, 2, 3]
lst2 = [2, 4, 5]
str1 = 'one'
str2 = 'two'

get_df(lst1, lst2, str1, str2)
````
